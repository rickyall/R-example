```r
options(repr.plot.width = 4,repr.plot.height = 3)
library('ggplot2')
data('diamonds')
```

# 第一章

```r
set.seed(1410)#让样本可重复
dsmall <- diamonds[sample(nrow(diamonds),100),]
```

```r
qplot(carat,price,data = diamonds)#画散点图
```

```r
qplot(log(carat),log(price), data = diamonds)#对数变换，取消指数
```

```r
qplot(carat,x * y * z, data = diamonds)#直接运行乘积，重量和体积的关系
```

```r
qplot(carat, price, data = dsmall, colour = color)#根据颜色添加颜色
```

```r
qplot(carat, price, data = dsmall, shape = cut)#根据形状添加切工信息
```

```r
qplot(carat,price,data = diamonds, alpha = I(1/10))#设置透明度
```

```r
#se = FALSE不加标准差,method = "loess"，使用局部回归的方法。
qplot(carat,price,data = dsmall,geom = c("point","smooth"))#添加拟合曲线和误差值
```

```r
qplot(color,price/carat,data = diamonds, geom = "jitter")#扰动点图
```

```r
qplot(color,price/carat,data = diamonds, geom = "boxplot")#象形胡须图
```

```r
qplot(carat,data=diamonds,geom="histogram",binwidth=0.1,colour=color)#直方图
```

```r
qplot(carat,data=diamonds,geom="density",fill = color)#密度曲线
```

```r
qplot(color,data=diamonds,geom="bar")#条形图
```

```r
qplot(color,data=diamonds,geom="bar",weight=carat)+scale_y_continuous("carat")#条形图
```
